Accesing Data with Pandas

```python
#show top 5 data
results.head()
```

```python
#show random data
results.sample(10)
```

```python
#accesing specific values (rows, column)
results.loc[0:3]
results.loc[[0,1,5], ["year", "athlete_id"]]
```

```python
#grabs the rows and column using index values
results.iloc[0:5, [0,2]]
```

```python
#change the table index to one of the values
results.index = results["athlete_id"]
results.loc[0:3]
```

```python
#modify data value
results.loc[1, "athlete_id"] = 10
```

```python
#other way to access specific values using at and iat (can only use 1 index)
results.at[0, "event"] 
results.iat[1, 3]
```

```python
#grabs column
results["event"]
#using dot for single values
results.athlete_id
```

```python
#sorting values
results.sort_values("athlete_id", ascending=False)
#sorting by 2 values
results.sort_values(["athlete_id","place"] , ascending=[0,1])
```
